```python
# library imports
import gurobipy as gp
from gurobipy import GRB
import numpy as np
from scipy.stats import poisson
import matplotlib.pyplot as plt
```

```python
shifts = ['Day', 'Evening', 'Night', ]
n_shifts = len(shifts)
expected_patient_counts = [56, 30, 50]
n_samples = np.arange(100)

day_patients = poisson.pmf(n_samples, expected_patient_counts[0])
evening_patients = poisson.pmf(n_samples, expected_patient_counts[1])
night_patients = poisson.pmf(n_samples, expected_patient_counts[2])

patient_data = [day_patients, evening_patients, night_patients]

fig, axes = plt.subplots(1, 3, figsize=(20, 5))
for c, ax in enumerate(axes):
  ax.bar(n_samples, patient_data[c], color='skyblue', edgecolor='black', alpha=0.7)
  ax.set_ylabel("Number of Patients")
  ax.set_title(f"{shifts[c]} Shift (λ = {expected_patient_counts[c]})")

plt.show()
```
